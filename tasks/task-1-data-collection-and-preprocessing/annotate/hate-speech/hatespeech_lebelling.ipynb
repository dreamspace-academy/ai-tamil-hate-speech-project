# The following script will map the excel lines with the words in stopwords defined in tamilhatespeech file

```python
import nltk
import pandas as pd
from datetime import datetime

import re    
import string     
import openpyxl 

hatewords = nltk.corpus.stopwords.words('tamilhatespeech')  
```

```python
# File processing
fp = "gender_videos_output.xlsx"

wb_obj = openpyxl.load_workbook(fp) 
sheet_obj = wb_obj.active 

row = sheet_obj.max_row
for i in range(1, row + 1):
    cell_obj = sheet_obj.cell( i , 1)
    text_data = cell_obj.value
    try:
        removed = tags_remove(text_data)
        n_list = list(map(lambda x: x.strip(), re.split(' ', removed)))
        processed_cell = sheet_obj.cell(row= i , column = 1) 
        processed_cell.value = " ".join(list(dict.fromkeys(n_list)))
        print(processed_cell.value)
    except:
        status_cell = sheet_obj.cell(row= i , column = 3) 
        status_cell.value = "exception"

    finally:
        print (i)

wb_obj.save(fp) 
print("Complete!!")

        
def tags_remove(text_data): 
    text = re.sub(r'#\S+', '', text_data)
    text = re.sub(r'https\S+', '', text)
    text = re.sub(r'@\S+', '', text)
    text = re.sub(r'"@\S+', '', text)
    return "".join([t for t in text if t not in string.punctuation])
```
